```python
import numpy as np
import matplotlib.pyplot as plt
import random


def normalize(A):
    scale_factor = A.max() - A.min()
    B = np.ones_like(A)*A.min()

    A = (A - B)/scale_factor
    return A

len = 200
a = [np.ones((1,10))]*len
b = [np.ones((1,10))]*len
rnd = np.random.random(len)
coh = []
for i in range(len):
    r = random.randint(0,9)
    b[i][0][r] += rnd[i]

    p = a[i]
    q = normalize(b[i])
    
    if np.array_equal(p,q) == False:
        coh.append(1/(np.linalg.norm(p-q,2)**2))
    else:
        coh.append(1)


x = list(range(len))
y = coh

plt.figure(figsize=(20,6), dpi = 500)
plt.gca().grid(True)
# plt.gca().set_facecolor((1, 1, 0.8))
plt.plot(x, y, label='old coherence formula',color='blue')

plt.title('Behavior of old coherence formula', fontsize=20)
plt.legend(loc="upper right",fontsize=15)
plt.ylabel("Coherence",fontsize=15)
plt.xlabel("Index",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig(fname= "D:\D_Drive\Github\Thesis\GraphSVX_graph_properties\graph_theoretic_properties\plots\old_coh_behavior")
plt.show()
```
